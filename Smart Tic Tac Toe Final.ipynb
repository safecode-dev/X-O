```python
class Player:
    def __init__(self, name, side):
        self.name = name
        self.vic = False
        self.side = side
        
    def __str__(self):
        return self.name
    
class Tic:
    def __init__(self):
        self.turn = None
        print("\nWelcome To the best Tic Tac Toe Game that Never Loses... :)\n\n")
        self.num_of_players = input("Please Choose an Option:\n1. Player VS Computer\n2. Player VS Player\n\n")
        if self.num_of_players == "2":
            name1 = input("First player Please Enter your name?\n")
            self.player1 = Player(name1, "X")
            name2 = input("Second player Please Enter your name?\n")
            self.player2 = Player(name2, "O")
            self.turn = self.player1
        else:
            name = input("Hello, Please Enter your name?\n")
            self.player2 = Player(name, "O")
        self.board = [[0, 0, 0],[0, 0, 0],[0, 0, 0]]
        self.mapping = f"""
In order to make a move please make sure you know the number of each cell:

 1 | 2 | 3
-----------
 4 | 5 | 6
-----------
 7 | 8 | 9

Good Luck!\n
"""
    def __str__(self):
        count = 0
        printable = ""
        for i in self.board:
            for j in i:
                count += 1
                if j == 0:
                    printable = printable + "|   "
                elif j == 1:
                    printable = printable + "| X "
                else:
                    printable = printable + "| O "
            if count % 3 == 0:
                printable = printable + "\n-------------\n"
        printable = printable[1:24] + "\n" + printable[28:51] + "\n" + printable[55:67]
        return "\n" + printable
    
    def xy_2_pos(self, x, y):
        dic = {(0, 0) : 1, (0, 1) : 2, (0, 2) : 3,
              (1, 0) : 4, (1, 1) : 5, (1, 2) : 6,
              (2, 0) : 7, (2, 1) : 8, (2, 2) : 9}     
        return dic[(x, y)]

    def pos_2_xy(self, pos):
        dic = {1 : (0, 0), 2 : (0, 1), 3 : (0, 2),
              4 : (1, 0), 5 : (1, 1), 6 : (1, 2),
              7 : (2, 0), 8 : (2, 1), 9 : (2, 2)}
        return dic[pos]
    
    def isempty(self, pos):
        x , y = self.pos_2_xy(pos)
        if self.board[x][y] == 0:
            return True
        return False
    
    def get_move(self):
        name = f"{self.turn}"
        pos = input(name + " It is your turn!!\nPlease choose a move (Enter a number bitween 1-9, make sure the place is empty): ")
        return self.input_check(pos)
        
    
    def input_check(self, pos):
        while True:
            while pos.isdigit() is False or int(pos) < 1 or int(pos) > 9:
                print("1 please Enter a number between 1 - 9...\n")
                pos = input("Please choose a move (Enter a number between 1-9, make sure the place is empty): ")
            if self.isempty(int(pos)) is False:
                print("2 This place is full, please choose another place\n")
                pos = input("Please choose a move (Enter a number between 1-9, make sure the place is empty): ")
                continue
            return int(pos)
    

    def vic_check_s(self, x, y, board):
        if self.vic_raw(x, board) or self.vic_col(y, board) or self.vic_diag1(board) or self.vic_diag2(board):
            self.vic = True
            return True
        return False
    
    def vic_check(self, pos, board):
        x, y = self.pos_2_xy(pos)
        return self.vic_check_s(x,y,board)

    def vic_raw(self, x, board):
        if self.board[x][0] and self.board[x][0] == self.board[x][1] == self.board[x][2]:
            return True
        return False

    def vic_col(self, y, board):
        if self.board[0][y] and self.board[0][y] == self.board[1][y] == self.board[2][y]:
            return True
        return False

    def vic_diag1(self, board):
        if self.board[0][0] and self.board[0][0] == self.board[1][1] == self.board[2][2]:
            return True
        return False

    def vic_diag2(self, board):
        if self.board[2][0] and self.board[2][0] == self.board[1][1] == self.board[0][2]:
            return True
        return False

    def isfullboard(self, board):
        for i in board:
            for j in i:
                if j == 0:
                    return False
        return True
    
    
    def empty_cells_list(self):
        empty_cells_list = []
        count_i = 0
        for i in self.board:
            count_j = 0
            for j in i:
                if j == 0:
                    empty_cells_list.append([count_i, count_j])
                count_j = count_j + 1
            count_i = count_i + 1
        return empty_cells_list
    
    
# Given a partially full board (might be full), returns the min-max move + score (a list of [x,y,score]). That is, if turn is 'X',
# returns the move that yields maximum for computer.
# And if turn is 'O', returns the move that yields minimum for computer.
    
    def min_max(self, board, turn):
        e_c_list = self.empty_cells_list()
        if len(e_c_list) == 0:
            return (-1, -1, 0)
        if turn == "X":
            score = -100
        else:
            score = 100
        while len(e_c_list):
            x, y = e_c_list[0][0], e_c_list[0][1]
            e_c_list.pop(0)
            if turn == "X":
                board[x][y] = 1
                if self.vic_check_s(x,y,board):
                    board[x][y] = 0
                    return (x,y,10)
                cur_x,cur_y,cur_score = self.min_max(board, 'O')
                if cur_score > score:
                    score = cur_score
                    chosen_x = x
                    chosen_y = y
                board[x][y] = 0
            else:
                board[x][y] = 2
                if self.vic_check_s(x,y,board):
                    board[x][y] = 0
                    return (x,y, -10)
                cur_x,cur_y,cur_score = self.min_max(board, 'X')
                if cur_score < score:
                    score = cur_score
                    chosen_x = x
                    chosen_y = y
                board[x][y] = 0
        return (chosen_x,chosen_y,score)
    
    def comp_move(self):
        print("It is My Turn:\n")
        x,y,max_score = self.min_max(self.board, "X")
        self.board[x][y] = 1
        return x, y


answer = "y"
while answer == "y" or answer == "yes":
    game = Tic()
    print(game.mapping)
    print(game)
    play = True
    if game.num_of_players == "2": # Player VS Player
        while play:
            game.turn = game.player1
            pos = game.get_move()
            x, y = game.pos_2_xy(pos)
            game.board[x][y] = 1
            print(game)
            if game.vic_check(pos, game.board):
                print(f'{game.player1} Glorious Victory ! ! ! :) ')
                break
            if game.isfullboard(game.board):
                play = False
                break

            game.turn = game.player2
            pos = game.get_move()
            x, y = game.pos_2_xy(pos)
            game.board[x][y] = 2
            print(game)
            if game.vic_check(pos, game.board):
                print(f'{game.player2} Glorious Victory ! ! ! :) ')
                break
            if game.isfullboard(game.board):
                play = False
                break
                
    else: # Comp VS Player
        print("I Go First :-)\n")
        while play:
            x, y = game.comp_move()
            print(game)
            pos = game.xy_2_pos(x, y)
            if game.vic_check(pos, game.board):
                print("I won!! Glorious Victory ! ! ! :) ")
                break
            if game.isfullboard(game.board):
                play = False
                break

            game.turn = game.player2
            pos = game.get_move()
            x, y = game.pos_2_xy(pos)
            game.board[x][y] = 2
            print(game)
            if game.vic_check(pos, game.board):
                print(f'{game.player2} Glorious Victory ! ! ! :) ')
                break
            if game.isfullboard(game.board):
                play = False
                break
            
    if not play:
        print("It's a TIE!! :-| ")
    answer = (input("Do you want another game? ")).lower()
```
